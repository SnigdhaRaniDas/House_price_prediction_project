```python
numeric_cols = data.select_dtypes(include=np.number).columns

for col in numeric_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(data[col], kde=True, bins=20, color='skyblue', edgecolor='black')
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

```python
corr_matrix = data.corr(numeric_only=True)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
scatter_visu=data.plot.scatter(x='Avg. Area Income',y='Price',figsize=(10,5),title='relation between avg area income and price')
plt.show()
```

```python
import matplotlib.pyplot as plt

# List of numerical features
features = [
    'Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population']

# Create scatter plots
plt.figure(figsize=(15,10))

for i, feature in enumerate(features):
    plt.subplot(2,3,i+1)
    plt.scatter(data[feature], data['Price'])
    plt.xlabel(feature)
    plt.ylabel('Price')
    plt.title(f'Price vs {feature}')

plt.tight_layout()
plt.show()
```

Key Findings / Observations
1. Most Important Features

From the scatter plots and correlation analysis, the features that appear most influential on house price are:

Avg. Area Income → Strong positive relationship with Price
Avg. Area Number of Rooms → Higher rooms generally increase price
Area Population → Moderate relationship with Price
Avg. Area House Age → Some positive effect
Avg. Area Number of Bedrooms → Weak relationship compared to rooms

Among these, Income and Number of Rooms are likely the strongest predictors.

2. Relationship Type (Linear or Non-Linear)

Most relationships appear approximately linear, especially:

Price vs Avg. Area Income

Price vs Avg. Area Number of Rooms

Some features may show slight non-linearity or scattered patterns, but overall the dataset is suitable for linear-based models.

3. Features Selected for Modeling

The following numerical features can be used for prediction:

Avg. Area Income
Avg. Area House Age
Avg. Area Number of Rooms
Avg. Area Number of Bedrooms
Area Population

The target variable is:

Price

The Address column is usually excluded because it is textual/categorical and difficult to use directly in basic regression models.

4. Suitable Machine Learning Models

a) Linear Regression

Good baseline model because many relationships are linear.

Best when:

Relationships are linear
Data has low complexity

b) Polynomial Regression

Useful if some relationships become curved/non-linear.

Best when:

Data shows non-linear trends

c) Ridge Regression

Helps reduce overfitting by adding regularization.

Best when:

Features are correlated
Model complexity increases

d) Lasso Regression

Useful for feature selection because it can reduce less important feature coefficients to zero.

Best when:

Some features may not contribute much

e) K-Nearest Neighbors (KNN)

Can capture non-linear patterns without assuming linearity.

Best when:

Data has complex local patterns

  Limitation:

Sensitive to scaling
Slower for large datasets
Final Observation

Since the dataset mostly shows linear relationships, models like:

Linear Regression
Ridge Regression
Lasso Regression

are likely to perform very well initially.

Polynomial Regression and KNN can also be tested to compare performance and detect hidden non-linear patterns.
